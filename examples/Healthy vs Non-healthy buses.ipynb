```python
#Time of Failure for every module in a particular bus
def tof(directory,bus):
    mtf={}
    directory_path=directory+bus
    #print(directory_path)
    if check_directory_existence(directory_path):
        print("Bus Folder Found")
        df_bus_dates = build_data_vis.sort_bus_by_date(directory, bus)
        no_of_files=df_bus_dates.shape[0]
        print ("Number of files in this bus are ", no_of_files)
        for i in range (1,17):
            mtf.update({f'{i}':[]})
        
        for module in range(1, 17):
            module_df = build_data_vis.build_module_average_df(directory, bus, module)
            for i in range (no_of_files-1):
    
                if (module_df.iloc[i,12]>module_df.iloc[i+1,12]):
                    mtf[f'{module}'].append(round(module_df.iloc[i,12]))
        #print(mtf)
        mtf_df=pd.DataFrame.from_dict(mtf,orient = 'index')
        if mtf_df.empty:
            print('The Bus is Healthy')
        else:
            print('Unhealty',mtf_df)

    else:
        print("The Bus Didn't Found")
```
